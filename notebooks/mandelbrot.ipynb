```python
import numpy as np
```

```python
N = 4000
a, b = -0.219, -0.22
c, d = -0.6995, -0.699
scale = .08
offsetx = 0.22
offsetz = 0.42
x = np.linspace(a + offsetx * (b - a), (b - a) * scale + offsetx * (b - a) + a, 2 * N)
z = np.linspace(c + offsetz * (d - c), (d - c) * scale + c + offsetz * (d - c), N)

xx, zz = np.meshgrid(x, z)
```

```python
a + offsetx * (b - a), (b - a) * scale + offsetx * (b - a) + a, c + offsetz * (d - c), (d - c) * scale + c + offsetz * (d - c)
```

```python
c = xx + 1j * zz

def draw(threshold = 120):
    render = np.zeros_like(xx)
    z = c
    for i in range(1, threshold + 1):
        z = z**2 + c
        render += (render == 0) * ((np.abs(z) > 4) * i)
    render[(render == 0)] = render.max()
    return render
```

```python
img = draw(1024)
```
