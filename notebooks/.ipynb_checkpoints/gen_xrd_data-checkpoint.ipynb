```python
def gen_train_and_test(n_basis, n_per_basis, prefix, same_basis = True, scale_type = 'shift'):
    basis = mk_basis_patterns(n_basis)
    x_train, y_train = mk_simdata(basis, n_per_basis, n_basis, scale_type = scale_type)
    if not same_basis:
        basis = mk_basis_patterns(n_basis)
    x_val, y_val = mk_simdata(basis, n_per_basis, n_basis, scale_type = scale_type)
    os.makedirs(prefix + '/train', exist_ok=True)
    os.makedirs(prefix + '/test', exist_ok=True)
    
    np.save("{}x.npy".format(prefix + '/train/'), x_train)
    np.save("{}y.npy".format(prefix + '/train/'), y_train)
    np.save("{}x.npy".format(prefix + '/test/'), x_val)
    np.save("{}y.npy".format(prefix + '/test/'), y_val)
    return x_train, y_train, x_val, y_val
```

```python
a, b, c, d = gen_train_and_test(100, 100, "data/100_member", same_basis = False)
```

```python
a, b, c, d = gen_train_and_test(10, 1000, "data/10_member_scale", same_basis = False, scale_type = 'scale')
```

```python
def condense2d(arr, newsize):
    extra = newsize - (len(arr) % newsize)
    arr = np.hstack((arr, np.zeros(extra)))
    return arr.reshape(newsize, -1).sum(axis = 1)
```
